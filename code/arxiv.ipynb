```python
ranked_auths_df = pd.read_pickle('final_pickles/flattened_chunks/final_ranked_authors_v3.pickle')
```

```python
ranked_auths_df_dict = {}
for i, r in ranked_auths_df.iterrows():
    ranked_auths_df_dict[r['authors']] = r['total_score']
```

```python
dblp_with_categories_df = pd.read_pickle('dblp_with_categories_final.pickle')
```

```python
dblp_with_categories_df.head()
```

```python
dblp_for_train = dblp_with_categories_df[['id', 'num_authors', 'total_year_citations', 'title_len', 'year', 'j-index', 'fos', 'authors']]
```

```python
dblp_for_train_year_stats = dblp_for_train.groupby(['year'])['id'].count().reset_index()
```

```python
dblp_for_train_year_stats.rename(columns={'id': 'num_year_papers'}, inplace=True)
```

```python
dblp_for_train = dblp_for_train.join(dblp_for_train_year_stats.set_index('year'), on='year', lsuffix='_l')
```

```python
dblp_for_train = dblp_for_train.reset_index()
dblp_for_train.head()
```

```python
def get_score(x):
    arr = []
    for a in x:
        if a in ranked_auths_df_dict:
            arr.append(ranked_auths_df_dict[a])
    return np.median(arr)

dblp_for_train['authors_score'] = dblp_for_train['authors'].apply(lambda x: get_score(x))
```

```python
dblp_for_train = dblp_for_train.reset_index()
```

```python
dblp_for_train.drop(columns=['index'], inplace=True)
```

```python
dblp_for_train.drop(columns=['level_0'], inplace=True)
```

```python
arxiv_fos_mapping = {}
```

```python
with open('arxiv_fields.csv','r') as arxiv_fields_f:
    fields = arxiv_fields_f.readlines()
    for line in fields:
        toks = line.split(':')
        arxiv_fos_mapping[toks[0].lower()] = toks[1].strip()
```

```python
arxiv_fos_mapping
```

```python
import json
ar_df['fos'] = ar_df['tag'].apply(lambda x: arxiv_fos_mapping.get(json.loads(x.replace('\'', '\"').replace("None", "\"None\""))[0]['term'].lower()), 'Other')
```

```python
ar_df = ar_df[ar_df['fos'] != 'Other']
ar_df = ar_df[ar_df['fos'] != None]
ar_df.head()
```

```python
cats = list(ar_df['fos'].unique())
cats.remove(None)
cats = set(cats)

dblp_for_train['fos'] = dblp_for_train['fos'].apply(lambda x: x if x in cats else 'Other')
dblp_for_train.head()
```

```python
dblp_for_train = dblp_for_train[dblp_for_train['fos'] != 'Other']
```

```python
import ast
def func(x):
    return [i['name'] for i in ast.literal_eval(x)]

ar_df['authors'] = ar_df['author'].apply(lambda x: func(x))
```

```python
# One hot encoding for categorical variables
category_col = 'fos'
labels = pd.get_dummies(dblp_for_train['fos'], prefix='is_fos')
dblp_for_train = dblp_for_train.drop('fos', axis=1)
dblp_for_train_encoded = dblp_for_train.join(labels)
dblp_for_train_encoded.head()
```

```python
dblp_for_train_encoded = dblp_for_train_encoded.dropna()
```

```python
scaler = MinMaxScaler()
scaler.fit(dblp_for_train_encoded[['authors_score', 'title_len', 'num_year_papers']])
```

```python
X = dblp_for_train_encoded[dblp_for_train_encoded.columns.values]
y = dblp_for_train_encoded['j-index']
X = X.drop(columns=['j-index'])
X_train, X_test, y_train, y_test = train_test_split(X, y)
start_time  = datetime.datetime.now()
gb_reg = GradientBoostingRegressor(n_estimators=100, max_depth=3)
gb_reg.fit(X_train, y_train)
end_time = datetime.datetime.now()
print('Training Done..., Time Cost: %d' % ((end_time - start_time).seconds))
```

```python
predicted_y = gb_reg.predict(X_test)
```
